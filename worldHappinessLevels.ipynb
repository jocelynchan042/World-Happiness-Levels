```python
# YOUR CODE HERE

# ML modeling
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Evaluation
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

<b>Task:</b> Use the rest of this notebook to carry out your project plan. 

You will:

1. Prepare your data for your model.
2. Fit your model to the training data and evaluate your model.
3. Improve your model's performance by performing model selection and/or feature selection techniques to find best model for your problem.

Add code cells below and populate the notebook with commentary, code, analyses, results, and figures as you see fit. 

```python
####################################
# Prepare data for model
####################################
# Create label and perform feature selection
y = df['Happiness Class']
selected_features=['Log GDP per capita', 
                   'Social support', 
                   'Healthy life expectancy at birth', 
                   'Freedom to make life choices', 
                   'Generosity',
                   'Perceptions of corruption',
                   'Positive affect',
                   'Negative affect',
                   'Confidence in national government',
                   'Democratic Quality',
                   'Delivery Quality']
X = df[selected_features]

# Split data into training and testing sets
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size=0.2, random_state=1234)

# Scale features for Logistic Regression
scaler = StandardScaler()
```

```python
############################################
# Fit model to training data
############################################
# Model 1: Default Logistic Regression

# Scale the data for Logistic Regression
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Default Logistic Regression
# multiclass needed for classification; solver used with multiclass; balanced helps mitigate class imbalance
defaultLR = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000, class_weight='balanced')
defaultLR.fit(X_train_scaled, y_train)


# Model 2: Random Forest Classifier

# Create and fit model to training data
# Start with n_estimators=100 and max_depth=10
rf_model = RandomForestClassifier(n_estimators=100, max_depth=10, class_weight='balanced', random_state=1234)
rf_model.fit(X_train, y_train)
```

```python
############################################
# Evaluate models on test set
############################################

# Model 1: Logistic Regression
y_pred_lr = defaultLR.predict(X_test_scaled)

# Evalute the accuracy
print("Logistic Regression:")

# 1) accuracy
print("LR accuracy score:", accuracy_score(y_test, y_pred_lr), "\n")

# 2) confusion matrix
class_labels = ['Low', 'Medium', 'High'] # for better clarity
c_mLR = pd.DataFrame(confusion_matrix(y_test, y_pred_lr), index=class_labels, columns=class_labels)
print("Default LR confusion matrix:\n", c_mLR, "\n")

# 3) classification report
print("LR classification report:\n", classification_report(y_test, y_pred_lr))


# Model 1: Random Forest
y_pred_rf = rf_model.predict(X_test)

# Evaluate the accuracy
print("\nRandom Forest:") # adding \n for easier formatting

# 1) accuracy
print("RF accuracy score:", accuracy_score(y_test, y_pred_rf), "\n")

# 2) confusion matrix
c_mRF = pd.DataFrame(confusion_matrix(y_test, y_pred_rf), index=class_labels, columns=class_labels)
print("RF confusion matrix:\n", c_mRF, "\n")

# 3) classification report
print("RF classification report:\n", classification_report(y_test, y_pred_rf))
```

Model Performance Analysis:

The Random Forest model is outperforming the Logistic Regression overall. 

1) Accuracy: RF has a higher accuracy score with 79.8% compared to LR's accuracy score of 73.7%.

2) Confusion Matrices: For the LR model, it tends to misclassify High as Medium and Medium as high. It predicts the Low class farily well with 216 out of 250 so around 86% accuracy for the Low class. The RF model is much better at predicting the High class, with 87% accuracy and less mistakes between the Medium and High classes. There is still some trouble with the Low vs High class with 69 incorrectly labeled as High.

3) Classification Reports:
   Precision: LR has the highest precision in the Medium class, then the High, and then the Low class. RF has the highest precision in the High class, then the Medium, then the Low.
   Recall: LR has highest recall in the High class, then Low, then Medium. RF has the highest recall in the Medium class, then High, and then Low.
   
   This shows that RF has higher precision for High and Medium classes whereas LR has a higher recall for High, but is still weaker overall.

```python
############################################
# Improve LR model performance 
# through GridSearchCV 
############################################

# Define parameter grid
param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['lbfgs', 'saga'], # lbfgs: multiclass; saga: L1 regularization
    'max_iter': [500, 1000],
    'class_weight': ['balanced']
}

# Initialize Logistic Regression model
lr = LogisticRegression(multi_class='multinomial', random_state=1234)

# Set up GridSearchCV
grid_searchLR = GridSearchCV(
    estimator=lr,
    param_grid=param_grid,
    cv=5,
    scoring='f1_weighted' # important for multiclass
)

# Fit GridSearchCV on scaled training data
grid_searchLR.fit(X_train_scaled, y_train)

# Show best hyperparameter
print("Best hyperparameter for LR:", grid_searchLR.best_params_)

# Predict with best model
y_pred_lr_tuned = grid_searchLR.predict(X_test_scaled)

# Evaluate tuned LR model
print("\nTuned Logistic Regression Performance:")
print("Accuracy:", accuracy_score(y_test, y_pred_lr_tuned))
print("\nConfusion Matrix:\n", pd.DataFrame(confusion_matrix(y_test, y_pred_lr_tuned),
                                          index=['Low', 'Medium', 'High'],
                                          columns=['Low', 'Medium', 'High']))
print("\nClassification Report:\n", classification_report(y_test, y_pred_lr_tuned))
```

According to these results, the best hyperparameters for Logistic Regression are C=10, class_weight=balanced, max_iter=500, and solver=lbfgs. 

In comparison to the default LR model from before, there is a slight improvement after tuning especially in terms of accuracy where the accuracy went up slightly. There are also some small improvements shown in the confusion matrix where the High class improved with more true positives than before though there is a slight drop in the Medium class. In the classification report, the tuned model only slightly improved recall for Medium class from 0.69 to 0.70.

Overall, this tuned version of the model improved model generalization slightly and these results from GridSearchCV also confirms that the original settings were already quite strong.

```python
############################################
# Improve RF model performance 
# through RandomizedSearchCV 
############################################

# Define parameter distribution
param_dist = {
    'n_estimators': [100, 200, 300, 400], # number of DT
    'max_depth': [5, 10, 15, 20, None],
    'min_samples_split': [2, 5, 10], # min number of samples required to split an internal ndoe
    'min_samples_leaf': [1, 2, 4], # min number of samples required to be a leaf node
    'class_weight': ['balanced']
}

# Initialize model
rf = RandomForestClassifier(random_state=1234)

# Set up RandomizedSearchCV
random_searchRF = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=20,               
    cv=5,
    scoring='f1_weighted',
    random_state=1234,
)

# Fit RandomizedSearchCV on unscaled training data
random_searchRF.fit(X_train, y_train)

# Predict with the best model
y_pred_rf_tuned = random_searchRF.predict(X_test)

# Show best hyperparameters
print("Best hyperparameters for RF:", random_searchRF.best_params_)

# Evaluate tuned Random Forest
print("\nTuned Random Forest Performance:")
print("Accuracy:", accuracy_score(y_test, y_pred_rf_tuned))
print("\nConfusion Matrix:\n", pd.DataFrame(confusion_matrix(y_test, y_pred_rf_tuned),
                                            index=['Low', 'Medium', 'High'],
                                            columns=['Low', 'Medium', 'High']))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf_tuned))
```

According to these results, the accuracy of the tuned Random Forest Classifier model decreased very slightly from 0.798 to 0.789 but class-level fairness improved which is more favorable for multi-class classification problems. 

For the confusion matrix, there is improvement in the Low and Medium classes whereas the High class is slightly worse. In the classification report, in the High class, there is a slight drop in precision but better recall in the tuned model. For the Low class, recall improved significantly from 0.66 to 0.78. There's a slight tradeoff in the Medium class but f1 remained strong. 

Overall, the tuned version of Random Forest shows better class-level performances and is more equitable across all classes. While the tuned Random Forest model has a slightly lower accuracy, it delivers better-balanced predictions across all classes. This trade-off is often desirable when class fairness and precision/recall for each category are more important than just raw accuracy.
